```python
import re
def extract_expandable_gb(x):
    if pd.isna(x):
        return None

    m = re.search(r'upto\s*(\d+)\s*(GB|TB)', x, re.I)
    if not m:
        return None

    size = int(m.group(1))
    unit = m.group(2).upper()

    return size * 1024 if unit == 'TB' else size
```

```python
df['memory_card_max_gb'] = df['card'].apply(extract_expandable_gb)
```

```python
df = df.drop(columns=['card'])
```

```python
df.loc[df['memory_card_supported'] == 0, 'memory_card_max_gb'] = 0
```

```python
cat_cols = ['os', 'sim_type', 'network_type', 'chipset','core_type', 'display_type', 'memory_card_type']
for c in cat_cols:
    df[c] = df[c].astype('category')
```
